```python
dataset_file = 'phospholipid summary.csv'
tasks = ['ApL(nm^2)', 'Thickness(nm)', 'K_a(Modulus)(mN/m)', 'Diffusion Coefficient(10^-8cm^2/s)']
featurizer_func = dc.feat.ConvMolFeaturizer()
loader = dc.data.CSVLoader(feature_field = 'smiles', featurizer=featurizer_func, tasks = tasks, id_field = 'Compound ID')
dataset = loader.create_dataset(dataset_file)
transformer = dc.trans.BalancingTransformer(dataset=dataset)
dataset = transformer.transform(dataset)
```

```python
model = dc.models.GraphConvModel(n_tasks=4, mode='regression')
metric = dc.metrics.Metric(dc.metrics.pearson_r2_score)
```

```python
model.fit(dataset, nb_epoch=1000)
print('training set score:', model.evaluate(dataset, [metric]))
```
